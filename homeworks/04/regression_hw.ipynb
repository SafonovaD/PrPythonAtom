##### 4. задание про кросс-валидацию 

выберите оптимальный параметр регуляризации на ridge с помощью кросс-валидации на трейне. померийте качество на тесте. Увеличелось ли оно на метриках?

```python
ridge = Ridge()
#alphas = np.logspace(-5,.5, 30)
alphas = [0.01, 0.05, 0.1, 1.0]
scores = list()
scores_std = list()

for alpha in alphas:
    ridge.alpha = alpha
    this_scores = cross_val_score(
        ridge, X_train_new, y_train, n_jobs=1, cv=5,
        scoring=make_scorer(r2_score, greater_is_better=True)
    )
    scores.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))
print(argmax(scores),scores[argmax(scores)],alphas[argmax(scores)])
#print(alphas)
```

```python
regr_ = Ridge()
regr_.fit(X_train_new, y_train)
regr_new = Ridge(alpha=alphas[argmax(scores)])
regr_new.fit(X_train_new, y_train)
```

```python
y_pred_ = regr_.predict(X_test_new)
y_pred_new = regr_new.predict(X_test_new)
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred_new)), r2_score(y_test, y_pred_new), 'Стало')

print(np.sqrt(mean_squared_error(y_test, y_pred_)), r2_score(y_test, y_pred_),'Было')
print('Качество немного увеличилось ')
```
